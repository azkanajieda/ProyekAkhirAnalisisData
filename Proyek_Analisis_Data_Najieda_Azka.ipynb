### Cleaning Data

```python
datetime_columns = ["order_delivered_carrier_date", "order_delivered_customer_date"]

for column in datetime_columns:
  orders_df[column] = pd.to_datetime(orders_df[column])
```

```python
orders_df.info()
```

**Insight:**
- Karena data sudah baik maka tidak perlu dilakukan Cleaning Data atau perbaikan data.
- Hanya dilakukan perubahan tipe data yang awalanya string (object) menjadi datetime agar dapat dliakukan analisis lamanya ekspedisi pengiriman

#### Explore orders_df

```python
orders_df.sample(5)
```

```python
delivery_time = orders_df["order_delivered_customer_date"] - orders_df["order_delivered_carrier_date"]
delivery_time = delivery_time.apply(lambda x: x.total_seconds())
orders_df["delivery_time"] = round(delivery_time/86400)
```

```python
orders_df.sample(5)
```

```python
orders_df.describe(include="all")
```

```python
orders_df.delivery_time.hist()
```

```python
customer_id_in_orders_df =  orders_df.customer_id
customers_df["status"] = customers_df["customer_id"].apply(lambda x: "Non Active" if x in customer_id_in_orders_df else "Active")
customers_df.sample(5)
```

```python
customers_df.groupby(by="status").customer_id.count()
```

```python
customers_df[customers_df.status == "Non Active"]
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana Performa Periode Pengiriman Ekspedisi di setiap State?

```python
plt.plot(customers_df["customer_state"], orders_df["delivery_time"], marker='o', linewidth=2, color="#72BCD4")
plt.title("Lamanya Pengiriman Berdasarkan State (Jam)", loc="center", fontsize=20)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

### Pertanyaan 2: Bagaimana Performa Periode Pengiriman Ekspedisi di setiap City?

```python
plt.plot(customers_df["customer_city"], orders_df["delivery_time"], marker='o', linewidth=2, color="#72BCD4")
plt.title("Lamanya Pengiriman Berdasarkan Stat (Jam)", loc="center", fontsize=20)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```
